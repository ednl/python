```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

random.seed()
viswanath = 1.1319882487943
points = 10_001
n = np.arange(points)
t = [1] * points
r = np.ones(points)
fib1 = fib2 = 1

for i in n[2:]:
    tmp = fib1
    if random.randint(0, 1):
        fib1 += fib2
    else:
        fib1 -= fib2
    fib2 = tmp
    t[i] = math.log(1 + abs(fib1))
    r[i] = math.exp(t[i] / i)
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(121)
plt.plot([1, points-1], [1, math.log(viswanath) * (points - 2)], color="r", linestyle="--")
plt.plot(n, t)
plt.title("Random Fibonacci sequence")
plt.xlabel("n")
plt.ylabel("log|t|")

plt.subplot(122)
plt.axhline(viswanath, color="r", linestyle="--")
plt.plot(n, r)
plt.title("Growth factor of the RFS")
plt.xlabel("n")
plt.ylabel("root(log|t|,n)")

plt.tight_layout()
plt.show()
```
