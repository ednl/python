```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

random.seed()
viswanath = 1.1319882487943
points = 10_000
x = np.arange(1, points + 1)
fib = np.ones(points, dtype=object)
logf = np.ones(points)
root = np.ones(points)

for n in range(2, points):
    if random.randint(0, 1):
        fib[n] = fib[n - 1] + fib[n - 2]
    else:
        fib[n] = fib[n - 1] - fib[n - 2]
    logf[n] = math.log(1 + abs(fib[n]))
    root[n] = math.exp(logf[n] / x[n])
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(121)
plt.plot([1, points], [1, math.log(viswanath) * points], color="r", linestyle="--")
plt.plot(x, logf)
plt.title("Random Fibonacci sequence")
plt.xlabel("n")
plt.ylabel(r"$\log{|t|}$")

plt.subplot(122)
plt.axhline(viswanath, color="r", linestyle="--")
plt.plot(x, root)
plt.title("Growth factor of the RFS")
plt.xlabel("n")
# Support for n-th root in LaTeX labels only available from matplotlib 3.4.0
# plt.ylabel(r"$\sqrt[n]{|t|}$")
plt.ylabel(r"$|t|^\frac{1}{n}$")

plt.tight_layout()
plt.show()
```
