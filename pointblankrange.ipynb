```python
from math import pi, sqrt, tan
import numpy as np
import matplotlib.pyplot as plt

g     = 9.81  # gravitation: 9.81 m/s^2
v     = 10    # slow throw: 10 m/s = 36 km/h = 22 mph
angle = 10    # degrees upward
tn    = tan(angle * pi / 180)  # tn=vy/vx => vy=tn.vx
vx    = v / sqrt(1 + tn * tn)  # vx^2+vy^2=v^2 => vx^2(1+tn^2)=v^2 => vx=v/sqrt(1+tn^2)
vy    = sqrt(v * v - vx * vx)

hero     = 1.80
villain  = 1.90
ratio    = 0.80  # shoulder height : overall height
target   = 0.45  # Torso Lethal Shot Placement rectangle of 450 Ã— 225 mm

t = np.linspace(0, 1, 81)
x = vx * t
y = hero * ratio + vy * t - 0.5 * g * np.square(t)

plt.subplots(figsize=(12, 12 * 2 / 5))
plt.axis([-0.1, 5.1, -0.1, 2.1])
plt.plot(x, y)
plt.grid(linestyle=':')
plt.vlines(np.arange(1, 6, 1), villain * ratio, villain * ratio - target, color='r', lw=5)
plt.axhline(y=hero * ratio, linestyle='--')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.show()
```
